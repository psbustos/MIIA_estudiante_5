```python
# YOUR CODE HERE
titulo = '# Tablero de control Covid 19'
#raise NotImplementedError()
```

```python
# YOUR CODE HERE
sel_1 = ["Casos", "Muertes"]
sel_2 = ["Histórico", "Últimos 7 días","Últimas 24 horas"]
seleccion1 = pn.widgets.Select(name='Casos o Muertes', options= sel_1)
seleccion2 = pn.widgets.Select(name='Periodicidad', options= sel_2)
#raise NotImplementedError()
```

```python
# YOUR CODE HERE
@pn.depends(seleccion1.param.value, seleccion2.param.value)

def funcion_interactiva(seleccion1, seleccion2): 
    datos = df 
    
    columna = "Cases - cumulative total"
    if(seleccion1 == "Casos" and seleccion2 == "Histórico"):
        columna = "Cases - cumulative total"
    elif(seleccion1 == "Casos" and seleccion2 == "Últimos 7 días"):
        columna = "Cases - newly reported in last 7 days"
    elif(seleccion1 == "Casos" and seleccion2 == "Últimas 24 horas"):
        columna = "Cases - newly reported in last 24 hours"
    elif(seleccion1 == "Muertes" and seleccion2 == "Histórico"):
        columna = "Deaths - cumulative total"
    elif(seleccion1 == "Muertes" and seleccion2 == "Últimos 7 días"):
        columna = "Deaths - newly reported in last 7 days"
    elif(seleccion1 == "Muertes" and seleccion2 == "Últimas 24 horas"):
        columna = "Deaths - newly reported in last 24 hours"
    else:
        columna
        
    df1 = df.sort_values(by=[columna], ascending=False).head(10)    
    fig1 = go.Figure(go.Bar( x=df1['Name'],y=df1[columna]))
    fig1.update_layout( title= str(seleccion1) + " " + str(seleccion2))
    
    return fig1

def funcion2():
    datos = df
    df2 = df.sort_values(by=["Deaths - cumulative total per 1 million population"], ascending=False).head(10)
    fig2 = go.Figure(go.Bar( y=df2['Name'],x=df2["Deaths - cumulative total per 1 million population"],orientation='h'))
    fig2.update_layout( title= "Muertes por millón de personas")
    return fig2

def funcion3():
    datos = df
    df3 =df.groupby(['WHO Region']).sum().sort_values(by=["Cases - cumulative total"], ascending=False).reset_index()
    fig3 = go.Figure(go.Bar( y=df3["WHO Region"],x=df3["Cases - cumulative total"],orientation='h'))
    fig3.update_layout( title= "Total de Casos por Región de OMS")
    return fig3
#raise NotImplementedError()
```

```python
# YOUR CODE HERE
tablero_control = pn.Column(pn.Row(pn.Column(titulo, seleccion1, seleccion2), funcion_interactiva),pn.Row(funcion2,funcion3))
#raise NotImplementedError()
```
