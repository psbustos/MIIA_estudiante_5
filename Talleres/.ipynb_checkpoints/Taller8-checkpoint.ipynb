```python
# YOUR CODE HERE
def valores_filtro_columna(columna):
    columna = columna.astype(str)       # Convertimos en str los valores de la columna que entra por parámetro.
    valores = columna.unique().tolist() # Guardamos una lista con los valores únicos de la columna que entra por parámetro.
    valores = sorted(valores)           # Organizamos la lista de valores únicos de manera ascedente.
    valores.insert(0, "Todos")          # Agregamos el elemento "Todos" en la primera posición de la lista.
    return valores 
#raise NotImplementedError()
```

### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado.
2. Número de niños en el hogar.
3. Número de adolescentes en el hogar.

```python
# YOUR CODE HERE
def filtrar(escolaridad, no_niños, no_adolescentes):
    DF = mercadeo
    # Los siguientes cuatro casos ajustan los registros a mostrar de acuerdo con la selección del usuario.
    if  escolaridad != "Todos" and no_niños != "Todos" and no_adolescentes != "Todos":
        display(DF[(DF["Education"] == escolaridad) & (DF["Kidhome"].astype(str) == no_niños)
                   & (DF["Teenhome"].astype(str) == no_adolescentes)])
    elif escolaridad == "Todos" and no_niños == "Todos" and no_adolescentes != "Todos": 
        display(DF[(DF["Teenhome"].astype(str) == no_adolescentes)])
    elif escolaridad == "Todos" and no_adolescentes == "Todos" and no_niños != "Todos":
        display(DF[(DF["Kidhome"].astype(str) == no_niños)])
    elif no_niños == "Todos" and no_adolescentes == "Todos" and escolaridad != "Todos":
        display(DF[(DF["Education"] == escolaridad)])
    elif escolaridad == "Todos" and no_niños != "Todos" and no_adolescentes != "Todos":
        display(DF[(DF["Teenhome"].astype(str) == no_adolescentes) & (DF["Kidhome"].astype(str) == no_niños)])
    elif escolaridad != "Todos" and no_niños != "Todos" and no_adolescentes == "Todos": 
        display(DF[(DF["Education"] == escolaridad) & (DF["Kidhome"].astype(str) == no_niños)])
    elif escolaridad != "Todos" and no_niños == "Todos" and no_adolescentes != "Todos":
        display(DF[(DF["Education"] == escolaridad) & (DF["Teenhome"].astype(str) == no_adolescentes)])
    else:
        display(DF)#raise NotImplementedError()
        
interactive_plot = interactive(
                                filtrar, 
                                escolaridad = valores_filtro_columna(mercadeo["Education"]),
                                no_niños = valores_filtro_columna(mercadeo["Kidhome"]),
                                no_adolescentes = valores_filtro_columna(mercadeo["Teenhome"])
                              )
display(interactive_plot)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
